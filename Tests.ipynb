```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load your Nifty 50 data into a pandas DataFrame
# Assuming you have a CSV file named 'nifty50_data.csv' with 'Date' and 'Close' columns
nifty_data = pd.read_csv('nifty_50_data.csv')
nifty_data['Date'] = pd.to_datetime(nifty_data['Date'])
nifty_data.set_index('Date', inplace=True)

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(nifty_data['Close'], color='blue')
plt.title('Nifty 50 Closing Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

### Stationary check 

#### ADF test

```python
# Augmented Dickey-Fuller Test for stationarity
adf_result = adfuller(nifty_data['Close'])
print('Augmented Dickey-Fuller Test:')
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value}')
```

#### KPSS Test

```python
# KPSS Test for stationarity
kpss_result = kpss(nifty_data['Close'])
print('\nKwiatkowski-Phillips-Schmidt-Shin (KPSS) Test:')
print('KPSS Statistic:', kpss_result[0])
print('p-value:', kpss_result[1])
print('Critical Values:')
for key, value in kpss_result[3].items():
    print(f'   {key}: {value}')
```
